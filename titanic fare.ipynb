```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['age'] = titanic_data['age'].fillna(np.mean(titanic_data['age']))
```

```python
titanic_data['deck'] = titanic_data['deck'].fillna(titanic_data['deck'].mode()[0])
```

```python
titanic_data['embarked'] = titanic_data['embarked'].fillna(titanic_data['embarked'].mode()[0])
```

```python
titanic_data['embark_town'] = titanic_data['embark_town'].fillna(titanic_data['embark_town'].mode()[0])
```

```python
titanic_data.isnull().sum()
```

```python
sns.distplot(titanic_data['age'], bins=10)
```

```python
sns.jointplot(x = "age", y = "fare", kind = "scatter", data = titanic_data)
```

```python
sns.rugplot(titanic_data['fare'])
```

```python
sns.barplot(x = "sex", y = "age", data = titanic_data, estimator = np.std)
```

```python
sns.swarmplot(x = "survived", y = "age", data = titanic_data, size = 4)
plt.title("Swarm plot of Age vs Survived ")
```

```python
# Exclude non-numeric columns from the DataFrame
numeric_columns = titanic_data.select_dtypes(include=['number'])

# Calculate the correlation matrix
tc = numeric_columns.corr()

# Plot the heatmap
sns.heatmap(tc, cmap="autumn")
plt.title("Correlation")
```
