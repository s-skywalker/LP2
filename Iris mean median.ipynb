```python
# Group by 'Species' and calculate mean for numeric variables
mean_values = df.groupby('Species').mean()

# Plot bar plot for mean values
mean_values.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Values of Numeric Variables by Species')
plt.xlabel('Species')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Numeric Variable')
plt.show()
```

```python
# Plotting SD

# Group by 'Species' and calculate standard deviation for numeric variables
std_values = df.groupby('Species').std()

# Plot bar plot for standard deviation values
std_values.plot(kind='bar', figsize=(10, 6))
plt.title('Standard Deviation of Numeric Variables by Species')
plt.xlabel('Species')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Numeric Variable')
plt.show()
```

```python
# Plotting Median 

# Group by 'Species' and calculate median for numeric variables
median_values = df.groupby('Species').median()

# Plot bar plot for median values
median_values.plot(kind='bar', figsize=(10, 6))
plt.title('Median of Numeric Variables by Species')
plt.xlabel('Species')
plt.ylabel('Median')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Numeric Variable')
plt.show()
```

```python
# Ploting Min 
# Group by 'Species' and calculate minimum values for numeric variables
min_values = df.groupby('Species').min()

# Plot bar plot for minimum values
min_values.plot(kind='bar', figsize=(10, 6))
plt.title('Minimum Values of Numeric Variables by Species')
plt.xlabel('Species')
plt.ylabel('Minimum Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Numeric Variable')
plt.show()
```

```python
# Plotting Max 
# Group by 'Species' and calculate maximum values for numeric variables
max_values = df.groupby('Species').max()

# Plot bar plot for maximum values
max_values.plot(kind='bar', figsize=(10, 6))
plt.title('Maximum Values of Numeric Variables by Species')
plt.xlabel('Species')
plt.ylabel('Maximum Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Numeric Variable')
plt.show()
```
