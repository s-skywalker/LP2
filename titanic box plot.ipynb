```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/titanic_data.csv')
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(np.mean(data['Age']))
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
sns.boxplot(x = data["Sex"], y = data["Age"])
```

```python
sns.boxplot(x = data["Sex"], y = data["Age"], hue = data["Survived"])
```

```python
sns.countplot(x = data["Survived"])
```

```python
data["Sex"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
plt.hist(data["Age"], bins = 10)
```

```python
sns.distplot(data["Age"])
```

```python
sns.scatterplot(x = data["Sex"], y = data["Age"], hue = data["Survived"])
```

```python
sns.barplot(x = data["Pclass"], y = data["Age"])
```

```python
sns.distplot(data[data["Survived"] == 0]["Age"], hist = False, color = "blue")
sns.distplot(data[data["Survived"] == 1]["Age"], hist = False, color = "orange")
```

```python
sns.heatmap(pd.crosstab(data["Pclass"], data["Survived"]))
```
